```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("whitegrid")


def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x):
    return np.maximum(0.01 * x, x)

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))


x = np.linspace(-5, 5, 1000)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(x, sigmoid(x))
ax.set_title("Funkcja Sigmoid")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(x, tanh(x))
ax.set_title("Funkcja Tanh")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(x, relu(x))
ax.set_title("Funkcja ReLU")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(x, leaky_relu(x))
ax.set_title("Funkcja Leaky ReLU")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(x, softmax(x))
ax.set_title("Funkcja Softmax")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```
