```python
import wandb
import pandas as pd
import pathlib

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")


data_dir = pathlib.Path("./results/quantization")
data_dir_phcd_paper = pathlib.Path("./results/phcd_paper")

arch_1 = pd.read_csv(data_dir / "optimalization_results_arch-1.csv")
arch_2 = pd.read_csv(data_dir / "optimalization_results_arch-2.csv")
arch_7 = pd.read_csv(data_dir / "optimalization_results_arch-7.csv")
arch_baseline = pd.read_csv(data_dir / "optimalization_results_arch-baseline.csv")

arch_1 = arch_1[['optimization_type', 'accuracy', 'compressed_disk_size']] 
arch_2 = arch_2[['optimization_type', 'accuracy', 'compressed_disk_size']]
arch_7 = arch_7[['optimization_type', 'accuracy', 'compressed_disk_size']]
arch_baseline = arch_baseline[['optimization_type', 'accuracy', 'compressed_disk_size']]

arch_1['Architektura'] = 'Architektura A - kwantyzacja float16'
arch_2['Architektura'] = 'Architektura B - kwantyzacja float16'
arch_7['Architektura'] = 'Architektura F - kwantyzacja float16'
arch_baseline['Architektura'] = 'Architektura Bazowa - kwantyzacja float16'

arch_1
```

# Other charts

```python
results_all = pd.read_excel(data_dir.parent / "Praca magisterska - wyniki.xlsx", sheet_name="Wszystkie litery")

results_all.columns = [
    "Architektura Bazowa",
    "Architektura A",
    "Architektura B",
    "Architektura C",
    "Architektura D",
    "Architektura E",
    "Architektura F",
]

models = ["architecture-1", "architecture-2",  "architecture-7", "architecture-baseline"]
model_labels = ["Architektura A", "Architektura B", "Architektura F", "Architektura Bazowa"]

results_all
```

```python
compressed_disk_size = pd.read_csv(data_dir_phcd_paper / "compressed_disk_size.csv")

for data in [compressed_disk_size]:
    data.columns = [ c.replace("Name: ", "") for c in data.columns ]

compressed_disk_size = compressed_disk_size.drop(columns=["Step"])

compressed_disk_size = compressed_disk_size[[c for c in compressed_disk_size.columns if "MIN" not in c and "MAX" not in c]]

compressed_disk_size = compressed_disk_size.T


summary_df = pd.DataFrame(columns=["model", "compressed_disk_size"])

for model, label in zip(models, model_labels):
    summary_df = summary_df.append({
        "model": label,
        "compressed_disk_size": compressed_disk_size.loc[f"{model} - compressed_disk_size"][0]
    }, ignore_index=True)

summary_df
```

```python
final_summary = summary_df

# sort like this: Baseline, A, B, C, D, E, F

final_summary = final_summary.sort_values(by=["model"], key=lambda x: x.map({
    "Architektura Bazowa": 0,
    "Architektura A": 1,
    "Architektura B": 2,
    "Architektura F": 6
}))

final_summary['compressed_disk_size'] = final_summary['compressed_disk_size'].astype(float)

# convert from bytes to MB
final_summary['compressed disk size (MB)'] = final_summary['compressed_disk_size'] / 1024 / 1024

final_summary
```

```python
formatted_final_summary = final_summary.copy()

formatted_final_summary["compressed disk size (MB)"] = formatted_final_summary["compressed disk size (MB)"].map(lambda x: f"{x:.2f}")
formatted_final_summary
```

```python
disk_size_comparison = quantized[['Architektura', 'compressed_disk_size']].copy()
# final_summary
disk_size_comparison['compressed disk size (MB)'] = disk_size_comparison['compressed_disk_size'] / 1024 / 1024
disk_size_comparison = disk_size_comparison.drop(columns=['compressed_disk_size'])


# group by "Architektura" and take median
disk_size_comparison = disk_size_comparison.groupby("Architektura").median().reset_index()
disk_size_comparison
```

```python
final_summary['Architektura'] = final_summary['model']

disk_size_comparison = pd.concat([final_summary[['Architektura', 'compressed disk size (MB)']], disk_size_comparison])
disk_size_comparison
```

```python
# create a plot based on ax.barh where bars are grouped into pairs and use the same color

pairs = [ ["Architektura A", "Architektura A - kwantyzacja float16"], ["Architektura B", "Architektura B - kwantyzacja float16"], ["Architektura F", "Architektura F - kwantyzacja float16"], ["Architektura Bazowa", "Architektura Bazowa - kwantyzacja float16"]]

fig, ax = plt.subplots(figsize=(15, 8))

models_cmap = plt.get_cmap("tab10")
models_cmap_list = [models_cmap(i) for i in range(len(model_labels))] 
models_cmap_list = models_cmap_list[:-1]
boxplot_cmap = models_cmap_list.copy()
boxplot_cmap.insert(3, "gray")

palette = boxplot_cmap#["indigo", "mediumseagreen", "royalblue",  "darkred"]
for i, pair in enumerate(pairs):
    #ax.barh(pair, disk_size_comparison[disk_size_comparison['Architektura'].isin(pair)]['compressed disk size (MB)'], color=palette[i])
    # color first bar in pair slightly darker
    ax.barh(pair[0], disk_size_comparison[disk_size_comparison['Architektura'].isin(pair)]['compressed disk size (MB)'].values[0], color=palette[i])
    ax.barh(pair[1], disk_size_comparison[disk_size_comparison['Architektura'].isin(pair)]['compressed disk size (MB)'].values[1], color=palette[i], alpha=0.5)


ax.set_xlabel("Rozmiar po kompresji (MB)")
ax.set_ylabel("Architektura")
ax.set_title("Rozmiar po kompresji dla poszczeglnych architektur przed i po kwantyzacji float16")

# save on desktop as kwantyzacja_kompresja.png 
plt.savefig("/home/wiktor/Pulpit/kwantyzacja_kompresja.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
compression_summary = pd.DataFrame(columns=["Architektura", "Pre quantization", "Post quantization", "pct reduction"])

for i, pair in enumerate(pairs):
    pre_quantization_size = disk_size_comparison[disk_size_comparison['Architektura'].isin(pair)]['compressed disk size (MB)'].values[0]
    post_quantization_size = disk_size_comparison[disk_size_comparison['Architektura'].isin(pair)]['compressed disk size (MB)'].values[1]
    pct_reduction = (pre_quantization_size - post_quantization_size) / pre_quantization_size * 100
    compression_summary = compression_summary.append({
        "Architektura": pair[0],
        "Pre quantization": f"{pre_quantization_size:.2f} MB",
        "Post quantization": f"{post_quantization_size:.2f} MB",
        "pct reduction": f"{pct_reduction:.2f}%"
    }, ignore_index=True)

compression_summary.columns = ["Architektura", "Przed kwantyzacj", "Po kwantyzacji", "Rnica"]
compression_summary.set_index("Architektura", inplace=True)
compression_summary
```

```python
accuracy_quantized = quantized[['Architektura', 'accuracy']].copy()

# group by "Architektura" and take median
#accuracy_quantized = accuracy_quantized.groupby("Architektura").median().reset_index()
accuracy_quantized.head()
```

```python
results_interesting = results_all[["Architektura A", "Architektura B", "Architektura F", "Architektura Bazowa"]].copy()

# show results_interesting as a table where each row contains 2 columns: "Architektura"EncodingWarning
results_interesting = results_interesting.melt(var_name="Architektura", value_name="accuracy")
results_interesting
```

```python
# concat with accuracy_quantized and convert accuracy to float
accuracy_quantized = pd.concat([accuracy_quantized, results_interesting])
accuracy_quantized['accuracy'] = accuracy_quantized['accuracy'].astype(float)
accuracy_quantized.sample(5)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

axes = axes.flatten()

boxplot_cmap = models_cmap_list.copy()
boxplot_cmap.insert(3, "gray")

quantization_palette = ["indigo", "mediumseagreen", "royalblue",  "darkred"]
for i, pair in enumerate(pairs):
    ax = axes[i]

    data = accuracy_quantized[accuracy_quantized['Architektura'].isin(pair)]

    # plot two boxplots, one for each element of pair, one is slightly darker than the other
    sns.boxplot(x="Architektura", y="accuracy", data=data, ax=ax, color=boxplot_cmap[i], order=pair)

    # set axis title as "Architektura X"
    #ax.set_title(pair[0])
    ax.set_xlabel(pair[0])

    # set y axis label to "Dokadno"

    # set x  labels to "Przed kwantyzacj" and "Po kwantyzacji"
    ax.set_xticklabels(["Przed kwantyzacj", "Po kwantyzacji"])

plt.savefig("/home/wiktor/Pulpit/kwantyzacja_dokladnosc.png", dpi=300, bbox_inches='tight')

plt.show()
```
